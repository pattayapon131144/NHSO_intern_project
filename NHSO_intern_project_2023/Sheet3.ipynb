```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
plt.rcParams['font.family'] = 'tahoma'
```

```python
disease_data = pd.read_csv('diease_data.csv')
disease_data
```

```python
disease_data_rpw_most100 = disease_data[disease_data['diease'] > 299]
#filter_rpw_most100 = disease_data_rpw_most100.sort_values(by='hosname', ascending=False)
disease_data_rpw_most100
```

```python
# Plotting the bar graph
plt.figure(figsize=(20, 6))
plt.bar(disease_data_rpw_most100['pdx'], disease_data_rpw_most100['diease'])
plt.xlabel('Hospital Name')
plt.ylabel('Number of Referrals (hcode_nb)')
plt.title('Top Ten Hospitals by Referrals')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

```python
disease_data_ampr = pd.read_csv('diease_data_ampr.csv')
disease_data_ampr
```

```python
filtered_data = disease_data_ampr[disease_data_ampr['amprhcode'] == 'เขตคลองสาน']
disease_filtered = filtered_data.sort_values(by='diease', ascending=False)
disease_filtered
```

```python
disease_data_ampr_rpw_most100 = disease_data_ampr[disease_data_ampr['diease'] > 50]
disease_filteredX = disease_data_ampr_rpw_most100.sort_values(by='diease', ascending=False)
disease_filteredX
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Set custom color palette
custom_palette = sb.color_palette("coolwarm", as_cmap=True)

heatmap_data = disease_data_ampr_rpw_most100.pivot_table(
    index='amprhcode', columns='pdx', values='diease', aggfunc='sum'
)
x_labels = heatmap_data.columns
y_labels = heatmap_data.index

plt.figure(figsize=(20, 15))
plt.yticks(ticks=range(len(y_labels)), labels=y_labels, fontsize=14, rotation=0)
plt.xticks(ticks=range(len(x_labels)), labels=x_labels, fontsize=10, rotation=45, ha='right')

# Use custom color palette and add grid lines
sb.heatmap(
    heatmap_data, annot=True, fmt='d', cmap=custom_palette,
    cbar=True, linewidths=0.5, linecolor='white', alpha=0.8,
    cbar_kws={'label': 'Number of Patients', 'orientation': 'vertical'}
)

# Set a darker background
plt.style.use('dark_background')

# Add lines around each cell in the heatmap
for _, spine in sb.heatmap(
    heatmap_data, annot=True, fmt='d', cmap=custom_palette,
    cbar=True, linewidths=0.5, linecolor='white', alpha=0.8
).spines.items():
    spine.set_visible(True)

# Set axis labels and title with Thai characters
plt.xlabel('เขตที่ผู้ป่วยเข้าใช้บริการ', fontsize=14)
plt.ylabel('เขตประจำของคนไข้', fontsize=14)
plt.title('Heatmap: จำนวนผู้ป่วยที่เข้าใช้บริการ (ที่มีจำนวนผู้ป่วย > 50)', fontsize=18)

# Increase font size of colorbar label
plt.colorbar().set_label('Number of Patients', rotation=270, labelpad=20, fontsize=12)

# Show the plot
plt.show()
```
